```python
!pip install yfinance
!pip install bs4
!pip install nbformat
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://pizza4ps.com/vn/career/")
html_data = response.text
soup = BeautifulSoup(html_data, 'html.parser')
print(soup)
```

```python
job_nodes = soup.findAll('li', class_='joball_listNode')
job_data = []
for node in job_nodes:
    link_tag = node.find('a', class_='joball_listLink')
    job_link = link_tag['href']
    job_data.append((job_link))
print("List jobs at Pizza 4P's")
for job_link in job_data:
    print(job_link)
```
